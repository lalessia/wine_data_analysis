```python
# importo le librerie necessarie 
import numpy as np
import pandas as pd
import os
```

```python
# Definire il percorso della cartella che contiene i file CSV
folder_path = "Capstone_Project_Gruppo1\wine for winery"

# Prendere con un ciclo for la lista di tutti i file CSV nella cartella
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

# Creo un DataFrame vuoto dove verranno concatenati tutti i file CSV
w_winery = pd.DataFrame()

# Concateno tutti i file CSV nella cartella in un unico DataFrame
for file in csv_files:
    file_path = os.path.join(folder_path, file)
    temp_df = pd.read_csv(file_path)
    w_winery = pd.concat([w_winery, temp_df], ignore_index=True)

# Stampa il DataFrame risultante
w_winery.head()
```
