# Vivino Scraping  Analysis

```python
# importo le librerie necessarie 
import numpy as np
import pandas as pd

%matplotlib inline 
import matplotlib   
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
vivino = pd.read_csv('Scraping_files_Vivino/Scraper.csv') 
vivino.head()
```

```python
vivino.info()
```

## Pulizia DF Vivino

```python
# elimino due colonne 
vivino = vivino.drop(['Unnamed: 0.1', 'Unnamed: 0'], axis=1)
```

```python
# Modifica dei Dtype del DF Vivino
```

```python
vivino['number ratings'] = vivino['number ratings'].replace("ratings","",regex=True).astype('int32')
```

```python
vivino['price'] = vivino['price'].replace("€","",regex=True).astype('float')
```

```python
# per aggiungere il simbolo € si può usare una lambda function, però questa funzione riporta il Dtype in object
''' vivino['price'] = vivino['price'].apply(lambda x: '€{:.2f}'.format(x)) '''
```

```python
vivino['bold_value'] = vivino['bold_value'].replace("%","",regex=True).astype('float')
```

```python
vivino['tannic value'] = vivino['tannic value'].replace("%","",regex=True).astype('float')
```

```python
vivino['sweet_value'] = vivino['sweet_value'].replace("%","",regex=True).astype('float')
```

```python
vivino['acidic_value'] = vivino['acidic_value'].replace("%","",regex=True).astype('float')
```

```python
vivino.info()
```

```python
vivino.head(3)
```

```python
# Riordino le colonne 
vivino = vivino[['product type', 'country of origin','name', 'grape', 'product rating', 'number ratings','price','bold_value','tannic value','sweet_value','acidic_value','food_pairing']]
```

```python
# Rinomino le colonne 
vivino.columns = ['Product type', 'Country of origin','Name', 'Grape', 'Product rating', 'Number ratings','Price €','Bold value %','Tannic value %','Sweet value %','Acidic value %','Food pairing']
```

```python
vivino.head()
```

## Analisi DF Vivino

```python
vivino.head()
```

```python
vivino.isna().sum()
```

```python
# Sostituisco solo i Nan della colonna Grape con Unrecognized grape
vivino['Grape'].fillna('Unrecognized grape', inplace=True)
```

```python
vivino.isna().sum()
```

```python
# Tipologia di vini più venduti
grape_ratings = vivino.groupby('Product type')['Number ratings'].agg(['sum', 'mean'])
grape_ratings.reset_index(drop=False)
```

```python
# Analisi di valutazione in stelline da 0.0 a 5.0 per Paese
country_star_ratings = vivino.groupby('Country of origin')['Product rating'].agg(['count', 'mean','min','max'])
country_star_ratings.reset_index(drop=False)
```

```python
#Paese con la Recensione più alta
best_country = country_star_ratings['max'].idxmax()
print("Paese con il conteggio delle Product rating maggiore è:", best_country)
```

```python
# Analisi del paese con piu vendite/recensioni
country_numb_ratings = vivino.groupby('Country of origin')['Number ratings'].sum()
country_numb_ratings.reset_index(drop=False)
```

```python
# Miglior Paese per Vendita
max_value = max(country_numb_ratings)
max_value = country_numb_ratings.nlargest(1)
max_value.reset_index(drop=False)
```

```python
# Tipi di Vino e Punteggio più alto
result_white = vivino.loc[(vivino['Product type'] == 'White wine') & (vivino['Product rating'] >= 4.5), ['Product type', 'Name','Product rating']]
result_white.sort_values(by='Product rating', ascending=False)
```

```python
# Vino e Punteggio più alto
result_red = vivino.loc[(vivino['Product type'] == 'Red wine') & (vivino['Product rating'] >= 4.5), ['Product type', 'Name','Product rating']]
result_red.sort_values(by='Product rating', ascending=False)
```

```python
# Vino e Punteggio più alto
result_rose = vivino.loc[(vivino['Product type'] == 'Rosé wine') & (vivino['Product rating'] >= 4.5), ['Product type', 'Name','Product rating']]
result_rose.sort_values(by='Product rating', ascending=False)
```

```python
# Vino e Punteggio più alto
result_sparkling = vivino.loc[(vivino['Product type'] == 'Sparkling wine') & (vivino['Product rating'] >= 4.5), ['Product type', 'Name','Product rating']]
result_sparkling.sort_values(by='Product rating', ascending=False)
```

```python
# Solo i 4 tipi di Vino che hanno il Rating piu alto
types_of_wine = ['Red wine', 'White wine', 'Sparkling wine', 'Rosé wine']

# Inizializza un dataframe vuoto
result_df = pd.DataFrame(columns=['Product type', 'Name', 'Product rating'])

# Per ogni tipo di vino, seleziona le righe corrispondenti, ordina per rating e seleziona il primo elemento
for wine_type in types_of_wine:
    filtered_df = vivino[vivino['Product type'] == wine_type]
    sorted_df = filtered_df.sort_values(by='Product rating', ascending=False)
    top_rated_wine = sorted_df.iloc[0][['Product type', 'Name', 'Product rating']]
    result_df = result_df.append(top_rated_wine, ignore_index=True)

# Visualizza i risultati
result_df.head()
```

```python
#Vini piu acquistati e recensiti
best_name_ratings = vivino.groupby(['Country of origin','Product type','Name','Grape'])['Number ratings'].sum()
best_name_ratings.reset_index(drop=False).sort_values(by='Number ratings',ascending=False).head()
```

```python
#Vini meno acquistati e recensiti
min_name_ratings = vivino.groupby(['Country of origin','Name','Grape'])['Number ratings'].sum()
min_name_ratings.reset_index(drop=False).sort_values(by='Number ratings').head()
```

```python
#I vini più Venduti con le valutazione (stelline) più alte
best_wine_region= vivino.groupby(['Product type','Name', 'Country of origin','Number ratings'])['Product rating'].mean().astype(int).reset_index(drop=False)
best_wine_region.sort_values(by='Number ratings', ascending=False).head()   
```

```python
#Groupby su Price per osservare quali sono i Tipi di vino più o meno costosi
vivino.groupby("Product type")["Price €"].agg(['count','sum','mean','min','max']).reset_index(drop=False)
```

```python
# Solo i 4 tipi di Vino che hanno il Price piu alto
types_of_wine = ['Red wine', 'White wine', 'Sparkling wine', 'Rosé wine']

# Inizializza un dataframe vuoto
result_df = pd.DataFrame(columns=['Product type', 'Name', 'Price €'])

# Per ogni tipo di vino, seleziona le righe corrispondenti, ordina per rating e seleziona il primo elemento
for wine_type in types_of_wine:
    filtered_df = vivino[vivino['Product type'] == wine_type]
    sorted_df = filtered_df.sort_values(by='Price €', ascending=False)
    top_rated_wine = sorted_df.iloc[0][['Product type', 'Name', 'Price €']]
    result_df = result_df.append(top_rated_wine, ignore_index=True)

# Visualizza i risultati
result_df.head()
```

```python
#Il vino più caro con relative Vendite(N. ratings)
best_wine_price= vivino.groupby(['Product type','Name', 'Country of origin','Number ratings','Product rating'])['Price €'].sum().astype(int).reset_index(drop=False)
best_wine_price.sort_values(by='Price €', ascending=False).head(1)
```

```python
#Il vino meno caro con relative Vendite(N. ratings)
best_wine_price= vivino.groupby(['Product type','Name', 'Country of origin','Number ratings','Product rating'])['Price €'].sum().astype(int).reset_index(drop=False)
best_wine_price.sort_values(by='Price €', ascending=True).head(1)
```

```python
# Paese con il tipo di Uva(grape) più comprata/recensita
best_grape_ratings = vivino.groupby(['Country of origin','Grape'])['Number ratings'].sum()
best_grape_ratings.reset_index(drop=False).sort_values(by='Number ratings',ascending=False).head(10)
```

```python
# Paese con il tipo di Uva(grape) meno comprata/recensita
min_grape_ratings = vivino.groupby(['Country of origin','Grape'])['Number ratings'].sum()
min_grape_ratings.reset_index(drop=False).sort_values(by='Number ratings').head(10)
```

## Analisi sulla Valutazione Rating/Price (Convenienza)

```python
# Nuova colonna che calcola il rapporto tra Rating e Price 
vivino["Rating Price Ratio"] = vivino["Product rating"] / vivino["Price €"]
```

```python
#Massimi valori della colonna "rating_price_ratio".
vivino.sort_values(by='Rating Price Ratio', ascending=False).head()
```

```python
# creo dei bin dove idati con valori compresi tra 0.0 e 0.20 saranno assegnati al primo bin, 
#i dati compresi tra 0.20 e 0.40 saranno assegnati al secondo bin, 
#i dati con valori compresi tra 0.40 e  saranno assegnati al terzo bin.
bins = [0.0, 0.2, 0.4, 1.0]
```

```python
# Creo una nuova colonna "Rating category" che mi permette di creare dei valori "Excellent", "Medium" o "Poor" in base al rating
vivino['Rating category'] = pd.cut(vivino['Rating Price Ratio'], bins=bins , labels=['Low', 'Medium', 'Excellent'])
vivino.head(10)   #i quantili megli farli sul Rating o sul rating price Ratio?????
```

```python
# Il mioglior Vino Rating - Price
sorted_data = vivino.sort_values(by='Rating Price Ratio', ascending=False)
best_wine_rp = sorted_data.iloc[0]
best_wine_rp
```

```python
# Il peggior Vino Rating - Price
sorted_data = vivino.sort_values(by='Rating Price Ratio')
worst_wine_rp = sorted_data.iloc[0]
worst_wine_rp
```

```python
# conto il numero di tipi di vino con rating "Excellent"
num_excellent = (vivino['Rating category'] == 'Excellent').sum()
print("Numero di vini con un rating 'Excellent':", num_excellent)
```

```python
# conta il numero di tipi di vino con rating "Medium"
num_medium = (vivino['Rating category'] == 'Medium').sum()
print("Numero di vini con un rating 'Medium':", num_medium)
```

```python
# conta il numero di tipi di vino con rating "Low"
num_low = (vivino['Rating category'] == 'Low').sum()
print("Numero di vini con un rating 'Low':", num_low)
```
